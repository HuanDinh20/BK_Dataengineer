## Big Data vs Small Data
To understand Big Data, we need to compare it with Small Data.
##### Small Data
<img src="img_2.png" width=1000>
1. Small Data is available in limited quantities that can be easily interpreted by humans with little or no digital processing. Sports scores and employee shift schedules are some examples of Small Data.
2. It is accumulated slowly and may or may not get updated continuously. Storing Small Data is relatively easy as it is mostly available in a structured format such as First and Last Name, Address, Gender, and so on, and It can be maintained within an enterprise’s own infrastructure or in a data center.

##### Big Data
1. Is generated in massive volumes and has little or no structure in which it is received.
Examples of semi-structured data include social media posts that could be images accompanied by hashtags, while unstructured data could
include medical records from millions of patients.
2. It is complex and requires specialized programs to interpret and make it available for human consumption. Not only is Big Data massive in size
3. it is collected continuously and grows exponentially in a short amount of time.
4. Big Data could be in any form including but not limited to text, images, audio, and videos.
5. Big Data is so voluminous that it is stored in the Cloud or on server farms set up specifically for this purpose.

## Bigdata Life Cycle
 It is a common misconception that Big Data refers to just large volumes of data. In reality, Big Data is the entire life cycle of working with large volumes of data.

 1. Big Data collection is initiated as a result of a business problem or requirement.
 2. As data is collected, it gets stored using a framework for distributed storage such as Hadoop HDFS.
 3. To make sense of all the data collected, Map and Reduce tasks and scripts create a data model to store it in a database
 4. This data model includes the various data entities (or objects), and the relationship and rules between these entities.
 5. After modeling, data is ready to be processed. Tools such as Apache Spark are used to produce meaningful information from the modeled data.
 6. Finally, the processed data is visualized and presented in a graphical format such as charts and graphs. This visualized data is then used for making meaningful business decisions and lead to new business cases, thereby creating a continuous life cycle.

## Characteristics of Big Data
The research firm Gartner defines Big Data as a high-volume, high-velocity, and high-variety information asset that demands cost-effective
and innovative tools for processing.

<img src="img_9.png" width=800>


### How huge Big Data volumes really are?

But big data can be even bigger. Do you know you can store over 11 million movies in 4k resolution in just one exabyte of space? Now visualize this… modern computers generally ship with 1- to 5-terabyte hard drives. One zettabyte contains a billion terabytes, while a yottabyte fits in one trillion terabytes? Doesn’t that boggle the mind? When we talk about Big Data, we traditionally talk about the four V’s of Big Data.

<img src="img_10.png" width=800>

### Four V’s of Big Data.
These are: Velocity, Volume, Variety, and Veracity

<img src="img_11.png" width=800>

1. **Velocity** is the speed at which data arrives.
2. **Volume** is the increase in the amount of data stored over time
3. **Variety** is the diversity of data. Many forms of data exist and need to be stored.
4. **Veracity** is the certainty of data.

With a large amount of data available, how will we know if the data collected is accurate or inaccurate? These four main components are used to describe the dimensions of Big Data.

### Velocity

<img src="img_12.png" width=800>

1. Velocity signifies that data is being generated extremely fast, and the process never stops.
2. Data must be processed quickly so that decisions can be made at the speed with which the data arrives.

#### Attributes
Velocity’s main attributes are:
1 .batch close to real-time, and
2. streaming.
####  Drivers
What are the drivers?
Definitely improving connectivity and hardware.

Just think about all the devices that are connected through the Internet today and all the super-fast response times. Big Data also supports upscaling of pre-calculated analysis.

### Volume

Volume is the increase in the amount of data stored. The amount of Big Data generated is vast compared to traditional data sources.

<img src="img_13.png" width=800>

#### Attributes
Volume attributes of Big Data are:
1. Petabytes,
2. Exa, and
3. Zetta, to name just a few.
#### Drivers
Typical drivers of volume in Big Data are:
1. the increase in data sources,
2. higher resolution sensors, and
3. scalable hardware infrastructure.

### Variety

<img src="img_14.png" width=800>

Variety is the diversity of the data. Data is generated by people and processes through the use of machines, from both inside and outside an organization. Some of the data is structured and semi-structured, but most is unstructured.
#### Attributes
The main attributes are
1. structure,
2. complexity,and
3. origin.

#### Drivers
Drivers of Variety in Big Data can be:
1. mobile technologies,
2. scalable hardware infrastructure,
3. resilience,
4. fault recovery, and
5. efficient storage and retrieval.

### Veracity

<img src="img_15.png" width=800>

Veracity is the quality, origin, and conformity to facts and accuracy of the data. This is because data comes from both within and outside an organization.

#### Attributes
Attributes include
1. consistency and
2. completeness,
3. integrity, and
4. ambiguity.

#### Drivers
Drivers of Veracity in Big Data are:
1. cost and the need for traceability -- Truy Xuất Nguồn Gốc --,
2. robust ingestion, and extract, transform, load (ETL) mechanisms.

Big Data has another V that must be considered. The fifth V of Big Data is Value.

### Value
 It is the outcome of making intelligent business decisions from leveraging the previous four V’s. The ultimate goal of an organization is to: produce value in the form of faster and smarter business decisions, increase efficient use of resources, and discover of new market opportunities. Big Data supports innovation and thus creates value.

### Summary
Big Data is the digital trace that gets generated in this digital era.



Big Data is a high-volume, high-velocity, and/or high-variety information asset that demands cost-effective and innovative tools for processing. The core features of Big Data are the 4 V’s:
1. Velocity,
2. Volume,
3. Variety, and
4. Veracity.

Big Data creates a fifth V, Value, when collected, stored, and processed correctly.
